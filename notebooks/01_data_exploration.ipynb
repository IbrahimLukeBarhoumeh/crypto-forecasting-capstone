```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from statsmodels.tsa.stattools import adfuller  # For stationarity checks

sns.set_theme(font_scale=1.1)

# ------------------------------------------------------------------------------------
# CONFIG
# ------------------------------------------------------------------------------------
DATA_FOLDER = '../data'  # Adjust if your .csv files are in a different folder
TICKERS = [
    "ADA","AVAX","BCH","BNB","BTC","DOGE","DOT","ETH","LEO","LINK",
    "LTC","MATIC","NEAR","SHIB","SOL","TON","TRX","UNI","XRP"
]

all_clean_dfs = {}

# ------------------------------------------------------------------------------------
# MAIN LOOP: LOAD & CLEAN EACH CSV
# ------------------------------------------------------------------------------------
for ticker in TICKERS:
    print(f"\n=== Exploring & Cleaning {ticker} ===")

   
    raw_csv = f"{ticker}.csv"  
    raw_path = os.path.join(DATA_FOLDER, raw_csv)

    if not os.path.exists(raw_path):
        print(f"[Skipping] File not found: {raw_path}")
        continue

    # 1) Load
    df = pd.read_csv(raw_path)

    
    df.rename(columns={
        "date":"Date",
        "open":"Open",
        "high":"High",
        "low":"Low",
        "close":"Close",
        "volume":"Volume"
        
    }, inplace=True)

    
    df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
    df.dropna(subset=['Date'], inplace=True)  # drop rows with invalid dates
    df.sort_values('Date', inplace=True)
    df.drop_duplicates(subset='Date', keep='first', inplace=True)
    df.reset_index(drop=True, inplace=True)

    
    df.ffill(inplace=True)
    df.dropna(subset=["Open","High","Low","Close","Volume"], inplace=True)

    
    df['Daily_Returns_%'] = df['Close'].pct_change() * 100.0
    df.dropna(inplace=True) 

    
    print("Final shape after cleaning:", df.shape)

   
    if 'Close' in df.columns:
        print(df[['Close']].describe())

        
        plt.figure(figsize=(10,4))
        plt.plot(df['Date'], df['Close'], label=f"{ticker} Close", color='blue')
        plt.title(f"{ticker} Price Over Time")
        plt.legend()
        plt.show()

        # B) Distribution of daily returns
        if 'Daily_Returns_%' in df.columns:
            plt.figure(figsize=(8,3))
            sns.histplot(df['Daily_Returns_%'], kde=True, bins=50)
            plt.title(f"{ticker} Daily Returns Distribution")
            plt.xlabel('Daily Returns (%)')
            plt.show()

        # C) Stationarity check (Augmented Dickey-Fuller)
        result = adfuller(df['Close'])
        pval = result[1]
        print(f"ADF p-value for {ticker}: {pval:.5f}")
        if pval < 0.05:
            print(f"{ticker} => likely stationary (p<0.05).")
        else:
            print(f"{ticker} => not stationary (p>0.05); differencing may help for ARIMA.")
    else:
        print(f"[Warning] {ticker} missing 'Close' column, skipping EDA charts...")

    # 8) Save the cleaned CSV
    cleaned_csv = f"{ticker}_cleaned.csv"
    cleaned_path = os.path.join(DATA_FOLDER, cleaned_csv)
    df.to_csv(cleaned_path, index=False)
    print(f"[Info] Saved cleaned data to: {cleaned_path}")

    # Store in dict for optional cross-coin correlation later
    all_clean_dfs[ticker] = df.copy()

# ------------------------------------------------------------------------------------
#  align each coin's daily close by date in one DataFrame, then do a heatmap.
# ------------------------------------------------------------------------------------
if not all_clean_dfs:
    print("\n[Info] No coin data found => skipping cross-coin correlation.")
else:
    
    common_dates = None
    for tkr, cdf in all_clean_dfs.items():
        these_dates = set(cdf['Date'])
        if common_dates is None:
            common_dates = these_dates
        else:
            common_dates = common_dates.intersection(these_dates)

    if not common_dates:
        print("\n[Info] No overlapping dates => skipping correlation heatmap.")
    else:
        common_dates = sorted(list(common_dates))
        cross_coin = pd.DataFrame({'Date': common_dates})

        
        for tkr, cdf in all_clean_dfs.items():
            sub = cdf[['Date','Close']].rename(columns={'Close': f"{tkr}_Close"})
            cross_coin = cross_coin.merge(sub, on='Date', how='left')

        cross_coin.dropna(inplace=True)
        if cross_coin.shape[1] <= 2:
            print("\n[Info] Not enough columns for correlation heatmap.")
        else:
            cross_coin.set_index('Date', inplace=True)
            plt.figure(figsize=(8,6))
            corr = cross_coin.corr()
            sns.heatmap(corr, cmap='coolwarm', annot=False)
            plt.title("Cross-Coin Correlation of Close Prices")
            plt.show()


```
