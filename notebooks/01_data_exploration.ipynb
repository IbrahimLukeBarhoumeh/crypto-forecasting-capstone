```python
# 01_data_exploration.ipynb

import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from statsmodels.tsa.stattools import adfuller  # For stationarity checks

sns.set_theme(font_scale=1.1)

DATA_FOLDER = '../data'

all_coins = [
    "Aave",
    "BinanceCoin",
    "Bitcoin",
    "Cardano",
    "ChainLink",
    "Cosmos",
    "CryptocomCoin",
    "Dogecoin",
    "EOS",
    "Ethereum",
    "Iota",
    "Litecoin",
    "Monero",
    "NEM",
    "Polkadot",
    "Solana",
    "Stellar",
    "Tether",
    "Tron",
    "Uniswap",
    "USDCoin",
    "WrappedBitcoin",
    "XRP"
]

# Store each coin's cleaned DataFrame in a dict for optional cross-coin correlation
all_clean_dfs = {}

for coin in all_coins:
    print(f"\n=== Exploring & Cleaning {coin} ===")
    raw_csv = f"coin_{coin}.csv"
    raw_path = os.path.join(DATA_FOLDER, raw_csv)
    if not os.path.exists(raw_path):
        print(f"Raw file not found: {raw_path}, skipping...")
        continue

    df = pd.read_csv(raw_path)

    # 1. Date -> datetime
    df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
    df.sort_values('Date', inplace=True)
    df.drop_duplicates(subset='Date', keep='first', inplace=True)
    df.reset_index(drop=True, inplace=True)

    # 2. Forward-fill
    df.ffill(inplace=True)

    # 3. Daily returns if 'Close' in columns
    if 'Close' in df.columns:
        df['Daily_Returns_%'] = df['Close'].pct_change() * 100

    # 4. Drop final NaNs
    df.dropna(inplace=True)
    print("Final shape after cleaning:", df.shape)

    # --- Basic EDA ---
    if 'Close' in df.columns:
        # A) Stats
        print(df[['Close']].describe())

        # B) Price over time
        plt.figure(figsize=(10,4))
        plt.plot(df['Date'], df['Close'], label=f"{coin} Close", color='blue')
        plt.title(f"{coin} Price Over Time")
        plt.legend()
        plt.show()

        # C) Distribution of daily returns
        if 'Daily_Returns_%' in df.columns:
            plt.figure(figsize=(8,3))
            sns.histplot(df['Daily_Returns_%'], kde=True, bins=50)
            plt.title(f"{coin} Daily Returns Distribution")
            plt.xlabel('Daily Returns (%)')
            plt.show()

        # D) Stationarity check (Augmented Dickey-Fuller)
        result = adfuller(df['Close'])
        print(f"ADF p-value for {coin}: {result[1]:.5f}")
        if result[1] < 0.05:
            print(f"{coin} likely stationary (p<0.05) under ADFuller.")
        else:
            print(f"{coin} not stationary (p>0.05), differencing may help for ARIMA.")
    else:
        print(f"{coin}: 'Close' column missing, skipping EDA charts...")

    # Save a copy
    cleaned_csv = f"coin_{coin}_cleaned.csv"
    cleaned_path = os.path.join(DATA_FOLDER, cleaned_csv)
    df.to_csv(cleaned_path, index=False)
    print(f"Saved cleaned data to: {cleaned_path}")

    # Optional: store in dictionary for cross-coin correlation
    all_clean_dfs[coin] = df.copy()

# === NEW CROSS-COIN CORRELATION (Optional) ===
# We'll align each coin's daily Close by Date in a single DataFrame
common_dates = None
for coin, cdf in all_clean_dfs.items():
    if common_dates is None:
        common_dates = set(cdf['Date'])
    else:
        common_dates = common_dates.intersection(set(cdf['Date']))

common_dates = sorted(list(common_dates))
cross_coin = pd.DataFrame({'Date': common_dates})
for coin, cdf in all_clean_dfs.items():
    # Merge on Date to get the coin's Close
    cdf2 = cdf[['Date','Close']].rename(columns={'Close': f"{coin}_Close"})
    cross_coin = cross_coin.merge(cdf2, on='Date', how='left')

cross_coin.dropna(inplace=True)
if len(cross_coin.columns) > 2:
    cross_coin.set_index('Date', inplace=True)
    plt.figure(figsize=(8,6))
    corr = cross_coin.corr()
    sns.heatmap(corr, cmap='coolwarm')
    plt.title("Cross-Coin Correlation of Close Prices")
    plt.show()
```
