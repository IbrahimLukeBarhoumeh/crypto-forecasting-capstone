```python


import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(font_scale=1.1)

DATA_FOLDER = '../data'


TICKERS = [
    "ADA","AVAX","BCH","BNB","BTC","DOGE","DOT","ETH","LEO","LINK",
    "LTC","MATIC","NEAR","SHIB","SOL","TON","TRX","UNI","XRP"
]

for ticker in TICKERS:
    print(f"\n=== Feature Engineering for {ticker} ===")

    # 1) The cleaned CSV is assumed to be e.g. "ADA_cleaned.csv"
    cleaned_csv = f"{ticker}_cleaned.csv"
    cleaned_path = os.path.join(DATA_FOLDER, cleaned_csv)
    if not os.path.exists(cleaned_path):
        print(f"[Skipping] File not found: {cleaned_path}")
        continue

    # cleaned CSV
    df = pd.read_csv(cleaned_path, parse_dates=['Date'])
    df.sort_values('Date', inplace=True)
    df.set_index('Date', inplace=True, drop=True)
    print("Initial shape:", df.shape)

    # Indicators

    # (A) Bollinger Bands (using 20-day rolling)
    df['MA_20'] = df['Close'].rolling(window=20).mean()
    rolling_std_20 = df['Close'].rolling(window=20).std()
    df['BB_upper'] = df['MA_20'] + (2 * rolling_std_20)
    df['BB_lower'] = df['MA_20'] - (2 * rolling_std_20)

    # (B) RSI(14)
    rsi_window = 14
    diff = df['Close'].diff()
    gain = np.where(diff > 0, diff, 0)
    loss = np.where(diff < 0, -diff, 0)
    avg_gain = pd.Series(gain, index=df.index).rolling(rsi_window).mean()
    avg_loss = pd.Series(loss, index=df.index).rolling(rsi_window).mean()
    rs = avg_gain / (avg_loss + 1e-9)
    df['RSI_14'] = 100.0 - (100.0 / (1.0 + rs))

    # (C) MACD (12,26), plus signal(9) & hist
    ema12 = df['Close'].ewm(span=12, adjust=False).mean()
    ema26 = df['Close'].ewm(span=26, adjust=False).mean()
    df['MACD'] = ema12 - ema26
    df['MACD_Signal'] = df['MACD'].ewm(span=9, adjust=False).mean()
    df['MACD_Hist'] = df['MACD'] - df['MACD_Signal']

    # (D) Stochastic(14)
    stoch_window = 14
    if 'Low' in df.columns and 'High' in df.columns:
        lowest_low_14 = df['Low'].rolling(stoch_window).min()
        highest_high_14 = df['High'].rolling(stoch_window).max()
        df['Stoch_%K'] = (df['Close'] - lowest_low_14) * 100.0 / (highest_high_14 - lowest_low_14 + 1e-9)
        df['Stoch_%D'] = df['Stoch_%K'].rolling(3).mean()
    else:
        print(f"[Info] Skipping Stoch for {ticker} => no 'Low'/'High' columns found.")

    # 4) Drop rows that became NaN from rolling calculations
    df.dropna(inplace=True)
    print("Shape after indicators & dropna:", df.shape)
    print("Final columns:", list(df.columns))

    #  Bollinger Plot
    plt.figure(figsize=(10,4))
    plt.plot(df.index, df['Close'], label='Close', color='blue')
    plt.plot(df.index, df['MA_20'], label='MA20', color='red', linestyle='--')
    plt.plot(df.index, df['BB_upper'], label='Upper BB', color='green', linestyle=':')
    plt.plot(df.index, df['BB_lower'], label='Lower BB', color='green', linestyle=':')
    plt.title(f"{ticker} with Bollinger Bands")
    plt.legend()
    plt.show()

    
    features_csv = f"{ticker}_features.csv"
    features_path = os.path.join(DATA_FOLDER, features_csv)
    df.to_csv(features_path)
    print(f"[Info] Saved feature-engineered data to: {features_path}")
```
